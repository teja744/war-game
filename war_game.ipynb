```python
import random
class Shuffle_cards:
    
    card = ['01-ace-spade','02-spade','03-spade','04-spade','05-spade','06-spade','07-spade','08-spade','09-spade','10-spade','11-jack-spade','12-queen-spade','13-king-spade','01-ace-heart','02-heart','03-heart','04-heart','05-heart','06-heart','07-heart','08-heart','09-heart','10-heart','11-jack-heart','12-queen-heart','13-king-heart','01-ace-diamond','02-diamond','03-diamond','04-diamond','05-diamond','06-diamond','07-diamond','08-diamond','09-diamond','10-diamond','11-jack-diamond','12-queen-diamond','13-king-diamond','01-ace-club','02-club','03-club','04-club','05-club','06-club','07-club','08-club','09-club','10-club','11-jack-club','12-queen-club','13-king-club']
    
    def __init__(self):
        pass
    def shuffle(self):
        shuf = input('-----PRESS S TO SHUFFLE-----')
        if shuf.upper() == 'S':
            random.shuffle(Shuffle_cards.card)    
        else:
            print('your cards are not shuffled'.upper())
    def variables(self):
        self.x = Shuffle_cards.card[0:26:]
        self.y = Shuffle_cards.card[26:]
        self.p1 = self.x
        self.p2 = self.y
        print(self.p1)
        print(self.p2)
```

```python
class Card(Shuffle_cards):
    
    def __init__(self):
        pass
        
    def name(self):
        self.player_1 = input('---enter player_1 name---')
        self.player_2 = input('---enter player_2 name---')
        
    def chance(self):
        import random
        if random.randint(0,1)==0:
            print('----' + self.player_1.upper() + ' got chance to start the game') 
        else:
            return print('----' + self.player_2.upper() + ' got chance to start the game')
        
    def start_the_game(self):
        import string
        alphabets=list(string.ascii_lowercase)   #to print all alphabets
        while True:
            self.start = input('---enter any alphabet to drop a card from your pack of cards--- '.upper())
            if self.start not in alphabets:
                print('---please enter only alphabets---')
            else:
                return 1
                break
    def option(self):
        self.name()
        self.chance()
        super().shuffle()
        super().variables()
```

```python
class Game_logic(Card):   #Card class is inherited here
    greater = []
    def __init__(self):
        pass
        
    def logic(self):
        if self.p1[0][0] > self.p2[0][0]:
            print(self.p1[0] + ' > ' + self.p2[0] )
            self.move()
            self.iterating()    
            self.clearing()
            self.v = input('-----click v to show your cards----')
            while self.v == 'v':
                print(self.p1)
                print(self.p2)
                break
            self.relogin()
        else:
            pass
        
    def equal(self):
        if self.p1[0][0] == self.p2[0][0]:
            if self.p1[0][1] == self.p2[0][1]:
                print(self.p1[0] + ' = ' + self.p2[0] )
                print('-----both cards have equal value----------')
                print('-----as per rules next cards of both players will hide-----'.upper())
                self.move()
                self.move()   #for moving hidden cards
                self.v = input('-----click v to show your cards-----')
                while self.v == 'v':
                    print(self.p1)
                    print(self.p2)
                    break
                self.relogin()   #again it will check from starting onwards
                
                
            elif self.p1[0][1] > self.p2[0][1]:
                print(self.p1[0] + ' > ' + self.p2[0] )
                self.move()
                self.iterating()
                self.v = input('-----click v to show your cards----------')
                while self.v == 'v':
                    print(self.p1)
                    print(self.p2)
                    break
                self.clearing()
                self.relogin()
                
            else:
                print(self.p1[0] + ' < ' + self.p2[0] )
                self.move()
                self.iterating_2()
                self.v = input('-----click v to show your cards-----')
                while self.v == 'v':
                    print(self.p1)
                    print(self.p2)
                    break
                self.clearing()
                self.relogin()
               
    def lower(self):
        self.option()
        self.permission()
        if self.p1[0][0] > self.p2[0][0]:
            self.logic()
        elif self.p1[0][0] == self.p2[0][0]:
            self.equal()
        else:
            print(self.p1[0] + ' < ' + self.p2[0] ,'lower')
            self.move()
            self.iterating_2()
            self.v = input('-----click v to show your cards-----')
            while self.v == 'v':
                print(self.p1)
                print(self.p2)
                break
            self.clearing()
            self.relogin()
       
    def move(self):
        self.greater.append(self.p1[0])
        self.greater.append(self.p2[0])
        self.p1.remove(self.p1[0])
        self.p2.remove(self.p2[0])
    
    def iterating(self):
        self.le = len(self.greater)
        self.lee = range(self.le)
        for k in self.lee:
            self.p1.append(self.greater[k])
    def iterating_2(self):
        self.fe = len(self.greater)
        self.fee = range(self.fe)
        for l in self.fee:
            self.p2.append(self.greater[l])
        
    def clearing(self):
        self.greater.clear()
    def permission(self):
         if self.start_the_game() == 1:
                pass
    def relogin(self):
        self.n = (input('-----enter n to next move------'))
        if self.n == 'n':
            if self.p1[0][0] > self.p2[0][0]:
                self.logic()
            elif self.p1[0][0] == self.p2[0][0]:
                self.equal()
            elif self.p1[0][0] < self.p2[0][0]:
                print(self.p1[0] + ' < ' + self.p2[0] , 'lower 2')
                self.move()
                self.iterating_2()
                self.v = input('-----click v to show your cards-----')
                while self.v == 'v':
                    print(self.p1)
                    print(self.p2)
                    break
                self.clearing()
                self.relogin()
            
            else:
                pass
        else:
            print('-----final result is-----'.upper())
            self.e = len(self.p1)
            self.ee = len(self.p2)
            print(self.e)
            print(self.ee)
            if self.e > self.ee:
                print('***** ' + self.player_1.upper() + ' WON THE GAME' + ' *****')
            elif self.e < self.ee:
                print('***** ' + self.player_2.upper() + ' WON THE GAME' + ' *****')
            else:
                print('*******DRAW MATCH*******')
```
